```python
# # # # # # # # #  Code for Task 1 # # # # # # # # #
Input_Dataset = pd.read_csv("dataset_group.csv",header=None) 
Input_Dataset.columns = ['Date','Transcation','Item']

#Sorting Dataset in alphabetical order
sorted_data=Input_Dataset.sort_values(by=['Transcation','Item'])
tran_len = len(Input_Dataset.Transcation.unique())

#Making The dataset intp list
listing_data = []
for i in range(1,tran_len):
        listing_data.append(sorted_data.Item[sorted_data.Transcation == i].tolist())

#writing modified data into dataset.csv        
final_dataset= pd.DataFrame(listing_data)
final_dataset.to_csv('dataset.csv', index=False, header=False)

dataset=pd.read_csv('dataset.csv', header=None)    
```
